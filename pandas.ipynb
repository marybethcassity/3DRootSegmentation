```python
df = pd.read_excel(r'C:\Users\Mary Beth Cassity\OneDrive - Auburn University\Grad\AgCypher Lab\RootAI\predictions\1viewsummary.xlsx',sheet_name = 'uncut')

df2 = pd.read_excel(r'C:\Users\Mary Beth Cassity\OneDrive - Auburn University\Grad\AgCypher Lab\RootAI\predictions\1viewsummary.xlsx',sheet_name = 'uncutDBSCAN')

df3 = pd.read_excel(r'C:\Users\Mary Beth Cassity\OneDrive - Auburn University\Grad\AgCypher Lab\RootAI\predictions\1viewsummary.xlsx',sheet_name = 'cut')

df4 = pd.read_excel(r'C:\Users\Mary Beth Cassity\OneDrive - Auburn University\Grad\AgCypher Lab\RootAI\predictions\1viewsummary.xlsx',sheet_name = 'cutDBSCAN')

fig, axes = plt.subplots(nrows=2, ncols=2)
plt.subplots_adjust(hspace=0.7)
fig.axes[0].set_ylabel('Dice')
fig.axes[2].set_ylabel('Dice')

ax = df.plot.bar(ax=axes[0,0], x = 'Percent', legend = False)
ax.set_ylim(0.6,1)
ax.set(xlabel="Percent of total dataset")
ax.set_title("Full plant")

ax2 = df2.plot.bar(ax=axes[0,1], x = 'Percent', legend = False)
ax2.set_ylim(0.6,1)
ax2.set(xlabel="Percent of total dataset")
ax2.set_title("Full plant with DBSCAN")

ax3 = df3.plot.bar(ax=axes[1,0], x = 'Percent', legend = False)
ax3.set_ylim(0.6,1)
ax3.set(xlabel="Percent of total dataset")
ax3.set_title("Root only")

ax4 = df4.plot.bar(ax=axes[1,1], x = 'Percent', legend = False)
ax4.set_ylim(0.6,1)
ax4.set(xlabel="Percent of total dataset")
ax4.set_title("Root only with DBSCAN")

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='center right', bbox_to_anchor=(1.1, 0.5))
```
